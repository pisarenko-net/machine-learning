## Correlations

Check if any features are highly correlated. If so we may need to combine them later.

```python
correlations = df_red.corr()

# Make the figsize 7 x 6
plt.figure(figsize=(7,6))
# Change color scheme
sns.set_style('white')
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Make the figsize 10 x 8
plt.figure(figsize=(18,12))
# Plot heatmap of correlations
sns.heatmap(correlations * 100, annot=True, fmt='.0f', mask=mask, cbar=False)
```

```python
correlations = df_white.corr()

# Make the figsize 7 x 6
plt.figure(figsize=(7,6))
# Change color scheme
sns.set_style('white')
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Make the figsize 10 x 8
plt.figure(figsize=(18,12))
# Plot heatmap of correlations
sns.heatmap(correlations * 100, annot=True, fmt='.0f', mask=mask, cbar=False)
```

## Fit regression to each feature

To gauge how much each feature contributes to the target variable.

```python
fig, axs = plt.subplots(nrows=6, ncols=2, figsize=(15,30))
for idx, feature in enumerate(df_red.columns.drop('quality')):
    sns.regplot(x=feature, y='quality', data=df_red, ax=axs[idx % 5][idx // 5])
```

```python
sns.regplot(x='alcohol', y='quality', data=df_red)
```

## Conclusion and next steps

There are no missing values in the dataset.

The target variable is categorical so this is a classification task.

Distributions of features differ between the red and white wines. Apart from the target variable, all features are floating point.

In both datasets there is medium positive correlation between total sulfur dioxide and free sulfur dioxide.

In the red wine set alcohol has a medium negative correlation to density. In the white set the correlation is stronger.

In the white wine set two features show high correlation: density and residual sugar.

The dataset is clean and nothing needs to be done in Data Cleaning. "Raw" data can be used as the ABT.

There is a significant imbalance between 'quality' classes. There are few extreme observations (9 and 3) and many more that fall into the middle. This means there isn't much data available to train on the extreme observations and by extension the results won't be great on outliers. Given the sparse classes they should be combined: bad (<=4), mediocre (5,6), good (>=6). The new feature should be used throughout the project.
