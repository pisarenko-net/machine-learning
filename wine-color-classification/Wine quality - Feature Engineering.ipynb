## Feature engineering

Create new variable 'color'. Drop variable 'quality'. Merge two datasets into one.

```python
# Import dependencies
from __future__ import print_function
# NumPy
import numpy as np
# Pandas
import pandas as pd
# Matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn
import seaborn as sns
```

## Load data

```python
df_red = pd.read_csv('winequality-red.csv', delimiter=';')
df_white = pd.read_csv('winequality-white.csv', delimiter=';')
```

## Drop quality variable

```python
df_red = df_red.drop('quality', axis=1)
df_red.head()
```

```python
df_white = df_white.drop('quality', axis=1)
df_white.head()
```

## Create color target variable

```python
df_red['color'] = 1
df_red
```

```python
df_white['color'] = 0
df_white
```

## Append datasets

```python
abt = df_red.append(df_white)
abt.head().append(abt.tail())
```

## Save ABT

```python
abt.to_csv('abt.csv', index=None)
```
